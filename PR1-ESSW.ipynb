```python


class PersonalityType(Fact):
    """Fact to hold answers for personality type questions."""
    pass

class PersonalityExpertSystem(KnowledgeEngine):
    def __init__(self):
        super().__init__()
        self.results = {}

    @Rule(PersonalityType(hypertimic=P(lambda x: x >= 5)))
    def hypertimic(self):
        self.results['Hypertimic'] = 'High probability'

    @Rule(PersonalityType(dysthymic=P(lambda x: x >= 5)))
    def dysthymic(self):
        self.results['Dysthymic'] = 'High probability'

    @Rule(PersonalityType(cyclothymic=P(lambda x: x >= 5)))
    def cyclothymic(self):
        self.results['Cyclothymic'] = 'High probability'

    @Rule(PersonalityType(excitable=P(lambda x: x >= 5)))
    def excitable(self):
        self.results['Excitable'] = 'High probability'

    @Rule(PersonalityType(stuck=P(lambda x: x >= 5)))
    def stuck(self):
        self.results['Stuck'] = 'High probability'

    @Rule(PersonalityType(pedantic=P(lambda x: x >= 5)))
    def pedantic(self):
        self.results['Pedantic'] = 'High probability'

    @Rule(PersonalityType(anxious=P(lambda x: x >= 5)))
    def anxious(self):
        self.results['Anxious'] = 'High probability'

    @Rule(PersonalityType(emotive=P(lambda x: x >= 5)))
    def emotive(self):
        self.results['Emotive'] = 'High probability'

    @Rule(PersonalityType(demonstrative=P(lambda x: x >= 5)))
    def demonstrative(self):
        self.results['Demonstrative'] = 'High probability'

    @Rule(PersonalityType(exalted=P(lambda x: x >= 5)))
    def exalted(self):
        self.results['Exalted'] = 'High probability'
```

```python
# Function to calculate scores for each personality type
def calculate_scores(answers):
    scores = {
        'hypertimic': sum(answers['Hypertimic']),
        'dysthymic': sum(answers['Dysthymic']),
        'cyclothymic': sum(answers['Cyclothymic']),
        'excitable': sum(answers['Excitable']),
        'stuck': sum(answers['Stuck']),
        'pedantic': sum(answers['Pedantic']),
        'anxious': sum(answers['Anxious']),
        'emotive': sum(answers['Emotive']),
        'demonstrative': sum(answers['Demonstrative']),
        'exalted': sum(answers['Exalted'])
    }
    return scores
```

```python
def run_expert_system(answers):
    scores = calculate_scores(answers)
    engine = PersonalityExpertSystem()
    engine.reset()
    engine.declare(PersonalityType(**scores))
    engine.run()
    return engine.results
```

```python

class EmotionRecognition(KnowledgeEngine):
    @DefFacts()
    def _initial_action(self):
        yield Fact(action="recognize_emotion")
        print("Система розпізнавання емоцій активована.")

    @Rule(Fact(action='recognize_emotion'), NOT(Fact(video=W())))
    def ask_for_video(self):
        self.declare(Fact(video=input("Введіть шлях до відео для аналізу: ")))

    @Rule(Fact(action='recognize_emotion'), Fact(video=MATCH.video))
    def preprocess_video(self, video):
        print(f"Обробка відео: {video}")
        self.declare(Fact(preprocessing="completed"))

    @Rule(Fact(preprocessing="completed"))
    def extract_features(self):
        print("Виділення ознак з відео...")
        self.declare(Fact(features="extracted"))

    @Rule(Fact(features="extracted"))
    def classify_emotion(self):
        print("Класифікація емоцій...")
        detected_emotion = "радість"  # Це значення можна змінити на основі реальної класифікації
        self.declare(Fact(emotion=detected_emotion))
        print(f"Виявлено емоцію: {detected_emotion}")

    @Rule(Fact(emotion=MATCH.emotion))
    def provide_result(self, emotion):
        print(f"Результат: емоція '{emotion}' успішно розпізнана.")
```

```python
# Запуск системи
engine = EmotionRecognition()
engine.reset()  # Ініціалізація фактів
engine.run()    # Запуск правил
```

```python
class HomeAutomation(Fact):
    """Факт для автоматизованої системи управління будинком"""
    pass

```

```python
class SmartHome(KnowledgeEngine):
    """Система продукційних правил для управління будинком"""

    @Rule(Fact(temperature=P(lambda x: x > 30)))
    def turn_on_ac(self):
        print("Температура перевищує 30°C. Увімкнення кондиціонера.")

    @Rule(Fact(temperature=P(lambda x: x < 18)))
    def turn_on_heater(self):
        print("Температура нижче 18°C. Увімкнення обігрівача.")

    @Rule(Fact(light=P(lambda x: x < 50)))
    def turn_on_lights(self):
        print("Освітленість низька. Увімкнення світла.")

    @Rule(Fact(light=P(lambda x: x > 200)))
    def turn_off_lights(self):
        print("Освітленість висока. Вимкнення світла.")

    @Rule(Fact(temperature=P(lambda x: 18 <= x <= 30), light=P(lambda x: 50 <= x <= 200)))
    def optimal_conditions(self):
        print("Умови оптимальні. Жодних дій не потрібно.")


```

```python
# Тестування системи
def test_smart_home():
    engine = SmartHome()
    engine.reset()

    # Додавання фактів
    engine.declare(Fact(temperature=25, light=60))


    # Запуск продукційної системи
    engine.run()

```
